```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10, 50)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, '--r') # --r = red color
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
plt.plot(x,y, '--b')#--b means blue color
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
plt.plot(x,y, '--r')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y, '--r')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = colours, s = sizes, alpha = .8) # intensity of color handle use alpha
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random data")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = colours, s = sizes, alpha = .8, marker='*')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random data")
```
